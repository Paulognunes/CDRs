```python
import folium
```

```python
mapa = folium.Map(location=[-18.9113, -48.2622], tales = "Stamen Terrain", zoom_start=12)
```

```python
for _, row in uberlandia_antenas.iterrows():
    folium.Marker([row['LAT'], row['LONG']], tooltip=str(row['ID'])).add_to(mapa)
    folium.CircleMarker([row['LAT'], row['LONG']], rad=50, fill=True).add_to(mapa)
```
